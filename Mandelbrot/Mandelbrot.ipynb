```python
# Imports

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Draw the Mandelbrot set

data = np.loadtxt('MandelbrotPoolsOnePyBlockSize16_128.csv', delimiter=',')
colors = data
plt.imshow(colors.T, extent=[-2,1,-1.5,1.5], cmap='grey')
plt.colorbar(label='Iterations')
plt.title('Mandelbrot Set')
plt.xlabel('Re')
plt.ylabel('Im')
plt.show()
```

```python
# Compare average calculation times vs grid sizes for all approaches

data = np.loadtxt('avgTime', delimiter=' ')
sizes = data[:,0]
times = data[:,1]

data_threads = np.loadtxt('avgTimeThreads', delimiter=' ')
sizes_threads = data_threads[:,0]
times_threads = data_threads[:,1]

data_pools_one = np.loadtxt('avgTimePoolsOneCoreBlocks', delimiter=' ')
sizes_pools_one = data_pools_one[:,0]
times_pools_one = data_pools_one[:,1]

data_pools_many = np.loadtxt('avgTimePoolsManyCoreBlocks', delimiter=' ')
sizes_pools_many = data_pools_many[:,0]
times_pools_many = data_pools_many[:,1]

plt.title('Average Time vs Size')
plt.xscale('log', base=10)
plt.yscale('log', base=10)
plt.xlabel('Pixels')
plt.ylabel('Time (s)')
plt.plot(sizes, times, marker='o', label='Sequential')
plt.plot(sizes_threads, times_threads, marker='o', label='Threads')
plt.plot(sizes_pools_one, times_pools_one, marker='o', label='Pools One')
plt.plot(sizes_pools_many, times_pools_many, marker='o', label='Pools Many')
plt.legend()
plt.show()
```
